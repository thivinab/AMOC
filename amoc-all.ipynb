```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import xarray as xr
import netCDF4 as nc
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import matplotlib.ticker as mticker

fn1 = './global.nc'
ds = xr.open_dataset(fn1)
ds = ds.sel(time=slice('1950-01-01', '2022-12-01'))
sst = ds.sst

g_clm = sst.groupby('time.month').mean('time');
g_an = sst.groupby('time.month') - g_clm;
g_an = g_an.mean(('lat','lon')); 

# Calculate the IQR (Interquartile Range)
Q1 = g_an.quantile(0.25)
Q3 = g_an.quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the annual_mean data to remove outliers
g_an = g_an.where((g_an >= lower_bound) & (g_an <= upper_bound), drop=True)

# Create a DataFrame with 'an_ds' and 'year-month' columns
gdf = pd.DataFrame({'g_an': g_an.values}, index=g_an.time.values)

# Reset the index to get 'year-month' as a separate column
gdf.reset_index(inplace=True)
gdf.rename(columns={'index': 'year-month'}, inplace=True)

#print(gdf)

```

```python
fn2 = './sg1950_2023.nc'
ds = xr.open_dataset(fn2)
ds = ds.sel(time=slice('1950-01-01', '2022-12-01'))
sst = ds.sst

clm = sst.groupby('time.month').mean('time');
an_ds = sst.groupby('time.month') - clm;
an_ds = an_ds.mean(('lat','lon')); 

# Calculate the IQR (Interquartile Range)
Q1 = an_ds.quantile(0.25)
Q3 = an_ds.quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the annual_mean data to remove outliers
an_ds = an_ds.where((an_ds >= lower_bound) & (an_ds <= upper_bound), drop=True)

#---------------------------------------------------------------------------------------------
amoc = an_ds - 2*g_an #AMOC Fingerprint

# Create a DataFrame with 'an_ds' and 'year-month' columns
df = pd.DataFrame({'amoc': amoc.values}, index=amoc.time.values)

# Reset the index to get 'year-month' as a separate column
df.reset_index(inplace=True)
df.rename(columns={'index': 'year-month'}, inplace=True)

################################################################################################


# Plotting the time series
plt.figure(figsize=(12, 6))
plt.plot(df['year-month'], df['amoc'], linestyle='-', color='black')
plt.xlabel('Time')
plt.ylabel('AMOC Fingerprint (deg C)')
#plt.title('Time Series of AMOC Fingerprint')

# Adding a trendline
x = np.arange(len(df))
z = np.polyfit(x, df['amoc'], 1)
p = np.poly1d(z)
plt.plot(df['year-month'], p(x), 'r--', label='Trendline')

plt.xticks(rotation=45)  # Rotate x-axis labels for readability
#plt.grid(True)
plt.tight_layout()

plt.title('c) AMOC Fingerprint',loc= 'left')


plt.show()
```
